```python
# Limpando dados faltante no DataFrame se houvesse
df1 = df1.dropna()
```

```python
# Verificando se há dados duplicados no DataFrame
df1.loc[df1.duplicated()]
```

```python
# Verificando os produtos, categorias e segmente mais vendidos por cidades, além de quantidade e valores, utilizando algumas variáveis do DataFrame
df1.loc[:, ['Product Name', 'Category', 'Segment', 'Country', 'City', 'Sales', 'Quantity']].sort_values(by='Sales', ascending=False).head(10)
```

```python
# Padronização de nomes de colunas
df1.columns = df1.columns.str.lower().str.replace(' ', '_').str.replace('[^a-zA-Z0-9_]', '', regex=True)
```

```python
# Verificação da Padronização de nomes de colunas
display(df1.columns)
```

```python
# Agregação de dados para cálculo de estatísticas de vendas
# Agrupando por categoria e calculando soma e média
stats_categoria = df1.groupby('category').agg({
    'sales': ['sum', 'mean', 'count'],
    'profit': ['sum', 'mean'],
    'quantity': 'sum'
})

# Renomeando colunas para facilitar a leitura
stats_categoria.columns = ['vendas_total', 'vendas_media', 'qtd_pedidos', 'lucro_total', 'lucro_medio', 'itens_vendidos']
print(stats_categoria.sort_values(by='vendas_total', ascending=False))
```

```python
# Agregação de dados para cálculo de estatísticas de vendas
resumo_vendas = df1['sales'].describe()
print(resumo_vendas)
```

```python
# Organização das tabelas para armazenamento em banco de dados
# Para a tabela de Clientes
df_clientes = df1[['customer_id', 'customer_name', 'segment']].drop_duplicates()

# Para a tabela de Produtos
df_produtos = df1[['product_id', 'product_name', 'category', 'subcategory']].drop_duplicates()

# Para a tabela de Localização
df_local = df1[['postal_code', 'city', 'state', 'region', 'country']].drop_duplicates()
```
